# Feminine and Masculine Cultures: 
## A comparison of the Nordics and United States

### Background:
   Geert Hofstede is a Dutch social psychologist who performed research on cross-culture groups and consequently, how national culture permeates and affects systems and organizations. In his book *Masculinity and Femininity: The Taboo Dimension of National Cultures*, he posits that countries fall on the spectrum of either having a masculine culture or a feminine one.
    
   His definition of culture in this study is more focused on “...broad patterns of thinking, feeling and acting…” (Hofstede et al. 5) and argue that these traits are deep-seated and are unlikely to change. On his ranking, the Nordic countries, comprised of Denmark, Finland, Iceland, Sweden, Norway, Greenland and the Faroe Islands, all rank as feminine, while the United States rank as being masculine.

### What is the difference?
__Feminine culture:__ stresses equality, quality, collaboration, compromise and consensus. These societies also have lower power distance, leading to flatter hierarchical structures. 

__Masculine culture:__ focuses on equity, competition and quantity. Masculine societies also have distinct social gender roles, where men are tough and assertive and women are tender. These societies have higher power distance and therefore more unequal distribution of power.

### Why is this relevant?
With another election around the corner, some politicians are already using the Nordics as an example of how we should model our social safety nets after. Some opponents are quick to argue that our cultures and systems in our country are too different from the Nordics and therefore, the policies are not adoptable. 

There is also that idea that our culture shapes our laws, then our laws perpetuate our culture. So if traits and views like strict gender roles are deep seated beliefs in our culture, will full equality for all be something that is achievable? 

## Hypothesis/Research Question:
To what extent is this masculine and feminine theory true in Nordic and American culture? Is it apparent in the language used in texts explaining how each is governed?

### Removing punctuations

```python
from string import punctuation 
print(punctuation)
```

### Tokenizing 

```python
nordic_token = word_tokenize(nordic_lower)
nordic_token
```

```python
# How many words?

print(len(nordic_token))
```

```python
# nordic_token is lowercase and in list. 

# arrange alphabetically.

nordic_token.sort()

# count frequencies 
from collections import Counter
nordic_freq = Counter(nordic_token)
nordic_freq
```

```python
# calling the pronouns individually 

nordic_freq['he']
nordic_freq['she']
nordic_freq['him']
nordic_freq['men']
nordic_freq['they']
```

### Removing stopwords

```python
import nltk
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
```

```python
# removing stop words from nordic_token 
from nltk.corpus import stopwords
```

```python
nordic_nostop = [word for word in nordic_token if word not in stopwords.words('english')]
nordic_nostop 
```

```python
# remove the strings that are just numbers because not needed.

nordic_nonumbers = [item for item in nordic_nostop if not item.isdigit()]
nordic_nonumbers
```

```python
from collections import Counter
nordic_nonumfreq = Counter(nordic_nonumbers)
nordic_nonumfreq
```

```python
# How many words without stopwords, no strings with numbers, and unique? 

print(len(nordic_nonumfreq))
```

```python
# top 70 most frequent words (this is when the word cooperation shows up)
nordic_nonumfreq.most_common(70)
```

```python
# turning frequency list into dataframe

nordic_df = pd.DataFrame(nordic_nonumfreq.most_common(30),
                        columns = ["Word", "Frequency"])
nordic_df
```

```python
nordic_barplot = nordic_df.plot.barh(x='Word', y='Frequency', 
                                  rot=0,
                                     color = "lightblue",
                                  title = "Nordic Word Frequencies",
                                  figsize = (11,8));
plt.xlabel("Frequency");
plt.ylabel("Word");
```
